```python
import os
import json
import numpy as np
```

```python
save_folder = "dataset"
name = "dataset"
input_name = os.path.join(save_folder, name + ".json")
output_name = os.path.join(save_folder, name + ".npy")
```

```python
with open(input_name, 'r') as f:
    data = json.load(f)
```

```python
data
```

```python
def key_convert(key):
    new_key = ""
    for i in range(0, len(key), 2):
        val = 2 * int(key[i]) + int(key[i+1])
        new_key += str(val)
    return new_key
```

```python
data_converted = {key_convert(key): val for key, val in data.items()}
```

```python
data_converted
```

```python
def string_to_array(string):
    ls = []
    for c in string:
        ls.append(int(c))
    return ls
```

```python
data_array = []

for result, count in data_converted.items():
    result = string_to_array(result)
    for i in range(count):
        data_array.append(result)
```

```python
data_array
```

```python
data_array = np.array(data_array)
```

```python
data_array_rescaled = data_array * 1/4 + 1/8
```

```python
data_array_rescaled
```

```python
rng = np.random.default_rng()
```

```python
data_array_spread = data_array_rescaled + rng.uniform(low = -1/8, high = 1/8, size = data_array_rescaled.shape)
```

```python
data_array_spread
```

```python
data_array_try_smear = data_array_spread + 0.1 * rng.standard_normal(size = data_array.shape)
```

```python
data_array_try_smear
```

```python
data_array_smeared = np.where((data_array_try_smear>=0) & (data_array_try_smear<=1), data_array_try_smear, data_array_spread) 
# make sure the output is between 0 and 1
```

```python
data_array_smeared
```

```python
np.save(output_name, data_array_smeared)
```
