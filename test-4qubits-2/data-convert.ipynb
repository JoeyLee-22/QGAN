```python
import os
import json
import numpy as np
```

```python
save_folder = "dataset"
name = "dataset"
input_name = os.path.join(save_folder, name + ".json")
output_name = os.path.join(save_folder, name + ".npy")
```

```python
with open(input_name, 'r') as f:
    data = json.load(f)
```

```python
data
```

```python
for result, count in data.items():
    print(result, count)
    print(type(result))
    print(type(count))
```

```python
def string_to_array(string):
    ls = []
    for c in string:
        ls.append(int(c))
    return ls
```

```python
data_array = []

for result, count in data.items():
    result = string_to_array(result)
    for i in range(count):
        data_array.append(result)
```

```python
data_array
```

```python
data_array = np.array(data_array)
```
